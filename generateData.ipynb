```python
import pandas as pd
import numpy as np
```

```python
def mockFormatCode (quantity):
    formatCodeSet = []
    formatCode = ""
    while len(formatCodeSet) < quantity:
        formatCode = "@TW"+pd.util.testing.rands(3).upper()
        formatCodeSet.append(formatCode)
        formatCodeSet = list(set(formatCodeSet))
    return formatCodeSet
```

```python
test01 = mockFormatCode(1000)
```

```python
testFollowers = mockFormatCode(30)
```

```python
analysisMatrix  = pd.DataFrame(index=[testFollowers],columns = [test01])
```

```python
analysisMatrix
```

```python
def generateRandomOnOff (quantity, probability):
    randomSet = []
    value=0
    for i in range(0,quantity):
        scale = np.random.rand() * 100
        if scale > probability:
            randomSet.append(1)
        else:
            randomSet.append(0)
    return randomSet
```

```python
matrixTransposed = analysisMatrix.transpose()
```

```python
matrixTransposed
```

```python
for value in matrixTransposed.columns:
    grouping = np.random.rand()* 200
    if grouping > 75:
        matrixTransposed[value] = generateRandomOnOff(30, 79)
    else:
        if grouping > 50:
            matrixTransposed[value] = generateRandomOnOff(30, 60)
        else:
            if grouping > 25:
                matrixTransposed[value] = generateRandomOnOff(30, 40)
            else:
                matrixTransposed[value] = generateRandomOnOff(30, 20)
```

```python
analysisMatrix = matrixTransposed.transpose()
```

```python
analysisMatrix.to_csv('DataSources/Twitter/Data.csv')
```
